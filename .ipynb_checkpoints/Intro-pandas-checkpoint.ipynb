```python
# some data are missing
car_sales_missing = pd.read_csv('car-sales-missing-data.csv')
car_sales_missing
```

```python
car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean())
```

```python
car_sales_missing
```

```python
#drop NaN row
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
# save to csv
car_sales_missing_dropped.to_csv('car-sales-missing-dropped.csv')
```

### Mainpulate data column

```python
car_sales_df
```

```python
# add seat column
car_sales_df['Seat'] = pd.Series([5,5,5])
car_sales_df
```

```python
# fill NaN
car_sales_df['Seat'].fillna(5, inplace=True)
car_sales_df
```

```python
# convert to int
car_sales_df['Seat'] = car_sales_df['Seat'].astype(int)
car_sales_df
```

```python
# add Fule Economy column assign with python list
fule_economy = [9.4, 7.6, 6.6, 9.9, 8.4, 9.3, 9.5, 5.5, 7.4, 7.7]
car_sales_df['Fule Per 100 KM'] = fule_economy
car_sales_df
```

```python
# add fule per 100 KM column
car_sales_df['Total Fule'] = car_sales_df['Odometer (KM)'] / 100 * car_sales_df['Fule Per 100 KM']
car_sales_df
```

```python
# round to second decimal for fule per 100 KM
car_sales_df['Total Fule'] = car_sales_df['Total Fule'].round(2)
car_sales_df
```

```python
# create column from single value
car_sales_df['Number Of Wheels'] = 4
car_sales_df
```

```python
car_sales_df['Passed Road Safety'] = True
car_sales_df
```

```python
car_sales_df.dtypes
```

```python
car_sales_df['certified'] = True
car_sales_df
```

```python
# drop a column
car_sales_df.drop('certified', axis=1)
```

```python
#shuffled 100% of data
car_sales_shuffled = car_sales_df.sample(frac=1)
car_sales_shuffled
```

```python
#shuffled 50% of data
car_sales_shuffled = car_sales_df.sample(frac=0.5)
car_sales_shuffled
```
